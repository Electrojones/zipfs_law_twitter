```python
import requests
import matplotlib.pyplot as plt

hashtags=["deutschland", "wir", "lol", "GNTM", "dasperfektedinner", "TRASH", "BR_quer"]
#hashtag=input("Enter your Hashtag!")


collection_of_counters=[]

for hashtag in hashtags:
    results=[]

    #get the twitter site
    response=requests.get("https://twitter.com/hashtag/"+hashtag+"?f=tweets&vertical=default&src=hash").text

    #init the list of the numbers to count
    numbers_to_count=list(range(0, 10))

    #count
    for number in numbers_to_count:
        results.append((response.count(str(number))))
        
    collection_of_counters.append(results)

fig, ax=plt.subplots(1)

for result in collection_of_counters:
    
    ax.plot(numbers_to_count, result, linewidth=0.5) 
ax.legend(hashtags,loc='upper right')
    #fig.show()
```
